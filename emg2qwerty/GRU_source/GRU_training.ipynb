#### Channel Ablation Training

```python
NUM_CHANNELS = 8

# Decided to use evenly-spaced channels to avoid biasing
channels = [int(round(i * 15 / (NUM_CHANNELS - 1))) for i in range(NUM_CHANNELS)]
IN_FEATURES = NUM_CHANNELS * 33 
CHANNELS_STR = "[" + ",".join(str(c) for c in channels) + "]"
```

```python
cmd = (
    f"python -m emg2qwerty.train"
    f" user=single_user"
    f" model=crnn"
    f" module.in_features={IN_FEATURES}"
    f" +module.electrode_channels={NUM_CHANNELS}"
    f" +module.select_channel_indices='{CHANNELS_STR}'"
    f" trainer.accelerator=gpu trainer.devices=1"
    f" trainer.max_epochs=40"
)
! {cmd}
```

#### Testing:

- Replace `Your_Path_to_Checkpoint` with your checkpoint path.

```python
# One-time compatibility patch for checkpoints missing architecture hparams
from pathlib import Path

train_py = Path("/content/emg2qwerty/emg2qwerty/train.py")
text = train_py.read_text()

old = """        module = module.load_from_checkpoint(
            config.checkpoint,
            optimizer=config.optimizer,
            lr_scheduler=config.lr_scheduler,
            decoder=config.decoder,
        )"""

new = """        module_kwargs = {k: v for k, v in config.module.items() if k != \"_target_\"}
        module = module.load_from_checkpoint(
            config.checkpoint,
            **module_kwargs,
            optimizer=config.optimizer,
            lr_scheduler=config.lr_scheduler,
            decoder=config.decoder,
        )"""

if old in text:
    train_py.write_text(text.replace(old, new))
    print("Patched train.py to pass module kwargs during checkpoint load.")
else:
    print("Expected pattern not found. train.py may already be patched.")
```
